## 5. An√°lisis de Valores Faltantes

Los valores faltantes (NaN, null) son un problema com√∫n. Debemos:
- Identificar qu√© variables los tienen
- Decidir qu√© hacer: ¬øimputar? ¬øeliminar?

```python
df.info()
```

```python
df.isnull().sum()
```

```python
# Contar valores nulos
missing = df.isnull().sum()
missing_pct = (df.isnull().sum() / len(df)) * 100

missing_df = pd.DataFrame({
    'Variable': missing.index,
    'Valores Faltantes': missing.values,
    'Porcentaje (%)': missing_pct.values
}).sort_values('Porcentaje (%)', ascending=False)

print("üîç An√°lisis de Valores Faltantes:\n")
missing_df[missing_df['Valores Faltantes'] > 0]
```

```python
# Visualizaci√≥n
fig, ax = plt.subplots(figsize=(12, 6))
missing_cols = missing_df[missing_df['Valores Faltantes'] > 0]

if len(missing_cols) > 0:
    ax.barh(missing_cols['Variable'], missing_cols['Porcentaje (%)'], color='coral')
    ax.set_xlabel('Porcentaje de Valores Faltantes (%)', fontsize=12)
    ax.set_title('Variables con Valores Faltantes', fontsize=14, fontweight='bold')
    ax.grid(axis='x', alpha=0.3)
    plt.tight_layout()
    plt.show()
else:
    print("‚úÖ No hay valores faltantes en el dataset")
```

## 6. Distribuci√≥n de Variables Num√©ricas

Vamos a ver c√≥mo se distribuyen las variables:
- ¬øSon normales?
- ¬øTienen sesgo?
- ¬øD√≥nde est√° concentrada la mayor√≠a de los datos?

```python
numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()
numeric_cols.remove('SeriousDlqin2yrs')

print(f"üìä Variables num√©ricas a analizar: {len(numeric_cols)}\n")

# Distribuciones
fig, axes = plt.subplots(3, 3, figsize=(16, 12))
axes = axes.ravel()

for idx, col in enumerate(numeric_cols):
    if idx < len(axes):
        # Histograma
        axes[idx].hist(df[col].dropna(), bins=50, color='steelblue', alpha=0.7, edgecolor='black')
        axes[idx].set_title(col, fontsize=11, fontweight='bold')
        axes[idx].set_xlabel('')
        axes[idx].grid(axis='y', alpha=0.3)
        
        # A√±adir estad√≠sticas
        mean_val = df[col].mean()
        median_val = df[col].median()
        axes[idx].axvline(mean_val, color='red', linestyle='--', linewidth=1.5, 
                         label=f'Media: {mean_val:.2f}')
        axes[idx].axvline(median_val, color='green', linestyle='--', linewidth=1.5, 
                         label=f'Mediana: {median_val:.2f}')
        axes[idx].legend(fontsize=8)

plt.tight_layout()
plt.show()
```
