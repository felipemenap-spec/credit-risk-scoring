## 8.5. Validaci√≥n Estad√≠stica: Point-Biserial Correlation

Nota t√©cnica: Cuando la variable dicot√≥mica est√° codificada como {0,1}, el coeficiente de correlaci√≥n de Pearson es algebraicamente equivalente al coeficiente Point-Biserial. En este an√°lisis se utiliza expl√≠citamente Point-Biserial para respetar la naturaleza binaria de la variable y evaluar formalmente la significancia estad√≠stica de la asociaci√≥n con cada variable continua.

```python
from scipy.stats import pointbiserialr

# Calcular Point-Biserial para todas las variables num√©ricas
print("üî¨ Point-Biserial Correlation (Variable Binaria vs Continua)\n")
print("=" * 90)
print(f"{'Variable':<45} {'r_pb':>12} {'p-value':>12} {'Significativa':>15}")
print("=" * 90)

pb_results = []

for col in numeric_cols:
    # Preparar datos (eliminar NaN)
    mask = ~(df['SeriousDlqin2yrs'].isna() | df[col].isna())
    x = df.loc[mask, 'SeriousDlqin2yrs'].astype(int).to_numpy()
    y = df.loc[mask, col].to_numpy()
    
    # Calcular Point-Biserial
    r_pb, p_value = pointbiserialr(x, y)
    
    # Determinar si es significativa
    significant = "‚úì S√≠ (p<0.05)" if p_value < 0.05 else "‚úó No"
    
    pb_results.append({
        'Variable': col,
        'r_pb': r_pb,
        'p_value': p_value,
        'Significativa': significant
    })
    
    print(f"{col[:43]:<45} {r_pb:>12.4f} {p_value:>12.6f} {significant:>15}")

print("=" * 90)

# Crear DataFrame ordenado por correlaci√≥n absoluta
pb_df = pd.DataFrame(pb_results)
pb_df['abs_r_pb'] = pb_df['r_pb'].abs()
pb_df = pb_df.sort_values('abs_r_pb', ascending=False)

print("\nüìä Top 5 Variables m√°s correlacionadas (por valor absoluto):\n")
print(pb_df[['Variable', 'r_pb', 'p_value']].head())
```

## 9. An√°lisis Bivariado: Variables vs Target

Comparemos las variables m√°s importantes entre clientes que hicieron default vs los que no.

```python
# An√°lisis de las top 3 variables m√°s correlacionadas
top_vars = target_corr.abs().nlargest(3).index.tolist()

fig, axes = plt.subplots(1, 3, figsize=(16, 5))

for idx, var in enumerate(top_vars):
    # Boxplot por target
    df_clean = df[[var, 'SeriousDlqin2yrs']].dropna()
    
    no_default = df_clean[df_clean['SeriousDlqin2yrs'] == 0][var]
    default = df_clean[df_clean['SeriousDlqin2yrs'] == 1][var]
    
    bp = axes[idx].boxplot([no_default, default], labels=['No Default', 'Default'],
                          patch_artist=True)
    
    # Colores
    for patch, color in zip(bp['boxes'], ['lightgreen', 'lightcoral']):
        patch.set_facecolor(color)
    
    axes[idx].set_title(f'{var} vs Target', fontsize=12, fontweight='bold')
    axes[idx].set_ylabel(var)
    axes[idx].grid(axis='y', alpha=0.3)

plt.tight_layout()
# plt.show()
```
